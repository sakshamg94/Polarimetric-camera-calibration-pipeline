```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

ni = 1
nt = 1.333
thI = np.linspace(0, np.pi/2, 200)
thT = np.arcsin(ni*np.sin(thI)/nt)

alpha = 0.5*(np.tan(thI-thT)/np.tan(thI+thT))**2
eta = 0.5*(np.sin(thI-thT)/np.sin(thI+thT))**2

DOLP = (alpha - eta)/(alpha + eta)
dDOLP = np.gradient(DOLP)

fig = plt.figure(figsize=(5, 5),  facecolor='w', edgecolor='k')
plt.plot(thI*180/np.pi, np.abs(DOLP),color='Indigo', linestyle='--', linewidth=3)
plt.axvline(x=53, linestyle='-', linewidth=3)
plt.xlabel("Incidence angle ($\\theta_i$) [deg]", fontsize=16)
plt.ylabel("DOLP($\\theta_i$)", fontsize = 16)
plt.xlim(0,90)
plt.grid()
plt.savefig("DOLP.png", format = "png")
plt.show()

fig = plt.figure(figsize=(5, 5), facecolor='w', edgecolor='k')
plt.plot(thI*180/np.pi, np.abs(dDOLP),color='red', linestyle='--', linewidth=3)
plt.axvline(x=37, linestyle='-', linewidth=3, color='green')
plt.axvline(x=53, linestyle='-', linewidth=3)
plt.xlabel("Incidence angle ($\\theta_i$) [deg]", fontsize=16)
plt.ylabel("$|\\frac{d~DOLP}{d~\\theta_i}|$", fontsize = 20)
plt.xlim(0,90)
plt.grid()
plt.savefig("grad_DOLP.png", format = "png")
plt.show()
```
